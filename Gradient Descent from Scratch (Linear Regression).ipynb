```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score,mean_squared_error
```

```python
df = pd.read_csv(r'C:\Users\prasa\OneDrive\Desktop\Regression\Data\airline cost new.csv')
```

```python
df.head()
```

```python
y=df.cost
X=df.num_pass
```

```python
plt.scatter(X,y)
plt.xlabel('Passengers')
plt.ylabel('Cost')
plt.show()
```

### Build gradient descent algorithm 

```python
m=10
c=1400
L=0.00001

epochs = 1000
n=len(X)

for i in range(epochs):
    y_pred = m*X+ c
    D_m = (-2/n)*sum(X*(y-y_pred)) ##slope of cost with respect to m
    D_c = (-2/n)*sum(y-y_pred) ##slope of cost with respect to c
    m = m - L*D_m  ## update m based on slope
    c = c - L*D_c  ## update c based on slope
    
    if (i%100)==0:
        plt.scatter(X,y)
        plt.plot(X,y_pred)
        plt.show()
        
print('m :',m,'c :',c)
```
