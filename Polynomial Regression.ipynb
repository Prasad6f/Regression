```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
from sklearn.datasets import load_boston
from sklearn.preprocessing import PolynomialFeatures
```

```python
boston = load_boston()
```

```python
bos = pd.DataFrame(boston.data, columns = boston.feature_names)
```

```python
bos['price']=boston.target
```

```python
y=bos['price']
X=bos.drop('price',axis=1)
#Xc = sm.add_constant(X)
```

```python
pf = PolynomialFeatures()

x = pf.fit_transform(X)
```

```python
list1 = pf.get_feature_names(X.columns)
x = pd.DataFrame(x,columns=list1)
```

```python
x.head()
```

```python
x.info(verbose=True)
```

```python
lin_reg = sm.OLS(y,x)
result=lin_reg.fit()
print(result.summary())
```

#### Backward elimination technique 

This is to drop the features, from huge datasets, that has no significance

```python
cols = x.columns.tolist()
pmax = 1
```

```python
while (len(cols)>0):
    X_1=x[cols]
    model=sm.OLS(y,X_1).fit()
    p=pd.Series(model.pvalues.values, index=cols)
    pmax=max(p)
    feature_max_p=p.idxmax()
    if (pmax > 0.05):
        cols.remove(feature_max_p)
    else:
        break
        
selected_features_BE = cols
print(selected_features_BE)
```

```python
x = x[selected_features_BE]
model=sm.OLS(y,x).fit()
print(model.summary())
```

The above prediction might leads to overfit as it contains full of multiplied features. So train & test split must be made.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
model=sm.OLS(y_train,X_train).fit()
#print(model.summary())
print(model.rsquared)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
y_pred = model.predict(X_test)
r2_score(y_test,y_pred)
```

Here the problem is overfitting. The model gives good accuracy for train set but not for test.

```python
mse = mean_squared_error(y_test,y_pred)
np.sqrt(mse)
```
